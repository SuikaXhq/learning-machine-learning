```python
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
import seaborn
import settings

plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=18)
# params= {'text.latex.preamble' : [r'\usepackage{amsmath}']}
plt.rcParams.update(params)
```

```python
# Data preparation
time = np.load('simulation/metrics/time.npy', allow_pickle=True).item()
print(time)

```

```python
# time plot
mark_size = 10
line_width = 4
legend_fontsize = 12
legend_markerscale = 0.8

fig = plt.figure()
ax = fig.add_subplot(311)
xtick = settings.get_sparsity()
ax.set_xticks(xtick)
ax.set_xlabel(r'degree of sparsity')
ax.set_ylabel('log-time(log-sec.)')
ax.set_title('time vs. sparsity', horizontalalignment='center', fontsize=20)

ax.plot(xtick, np.log10(time['PCA']['sparse']), 's-', ms=mark_size, linewidth=line_width, label=r'PCA')

ax.plot(xtick, np.log10(time['SPCA']['sparse']),'^:', ms=mark_size, linewidth=line_width, label=r'SPCA')

ax.plot(xtick, np.log10(time['dPCA']['sparse']), 'v--', ms=mark_size, linewidth=line_width, label=r'dPCA')

ax.plot(xtick, np.log10(time['L2_dPCA']['sparse']), 'v--', ms=mark_size, linewidth=line_width, label=r'L2-dPCA')

ax.plot(xtick, np.log10(time['EE-dPCA']['sparse']), 'v--', ms=mark_size, linewidth=line_width, label=r'EEdPCA')

ax.legend(loc='upper right', frameon=False,fontsize=legend_fontsize,labelspacing=0,markerscale=legend_markerscale,borderaxespad=0)
```
