```
# calculate F1 score
F1_S = pd.DataFrame(columns=['N', 'P', 'S','D', 'T', 'sparsity', 'dPCA', 'L2_dPCA', 'EEdPCA'])
# S
def calculate_F1(df_dpca, df_l2dpca, df_eedpca):
    F1_dpca = df_dpca['TP']/(df_dpca['TP'] + 0.5*(df_dpca['FP']+df_dpca['FN']))
    F1_l2dpca = df_l2dpca['TP']/(df_l2dpca['TP'] + 0.5*(df_l2dpca['FP']+df_l2dpca['FN']))
    F1_eedpca = df_eedpca['TP']/(df_eedpca['TP'] + 0.5*(df_eedpca['FP']+df_eedpca['FN']))
    F1 = pd.concat([df_dpca.iloc[:, :7], F1_dpca, F1_l2dpca, F1_eedpca], axis=1).iloc[:, 1:]
    F1.columns = ['N', 'P', 'S','D', 'T', 'sparsity', 'dPCA', 'L2_dPCA', 'EEdPCA']
    return F1
F1_D = calculate_F1(D_dPCA, D_L2_dPCA, D_EEdPCA)
F1_S = calculate_F1(S_dPCA, S_L2_dPCA, S_EEdPCA)
F1_T = calculate_F1(T_dPCA, T_L2_dPCA, T_EEdPCA)

print(F1_S)
print(F1_D)
print(F1_T)

F1_D.to_csv('simulation/metrics/F1_D.csv', index=None)
F1_S.to_csv('simulation/metrics/F1_S.csv', index=None)
F1_T.to_csv('simulation/metrics/F1_T.csv', index=None)
```

```
def draw_marginEVs(N, P, T, S, D, sparsity):
    name1 = ['pca', 'dpca', 'l2dpca', 'eedpca']
    name2 = ['PCA', 'dPCA', 'L2dPCA', 'EEdPCA']
    
    fig = plt.figure(figsize=(12,10))
    n_components = 10
    width = 0.75
    xtick = np.arange(1,n_components+1)
    mark_size = 10
    line_width = 1
    legend_fontsize = 14
    legend_markerscale = 0.8
    t_color = 'darkgray'
    s_color = 'royalblue'
    d_color = 'indianred'
    sd_color = 'mediumpurple'
    alpha = 1

    for method in range(4):
        EV = np.load('simulation/metrics/N{}_P{}_S{}_D{}_T{}_s{}/EV_{}.npy'.format(N, P, S, D, T, sparsity, name1[method]), allow_pickle=True).item()
        ax = fig.add_subplot(2,2,method+1)
        ax.set_title('Demixing Performance of {}\n\\Large($n={},p={},S={},H={},T={},s\%={}\%$)'.format(name2[method], N, P, S, D, T, int(sparsity*100)))
        ax.set_ylabel('Explained Variance (\%)')
        ax.set_xlabel('Component')
        ax.set_xticks(xtick)

        #########################################################################################################################################################

        mEV_t = np.asarray([EV['marginVar'][i]['t'] for i in range(n_components)])
        mEV_s = np.asarray([EV['marginVar'][i]['s']+EV['marginVar'][i]['st'] for i in range(n_components)])
        mEV_d = np.asarray([EV['marginVar'][i]['d']+EV['marginVar'][i]['dt'] for i in range(n_components)])
        mEV_sd = np.asarray([EV['marginVar'][i]['sd']+EV['marginVar'][i]['sdt'] for i in range(n_components)])

        
        bar_t = plt.bar(xtick, mEV_t, width, linewidth=line_width, color=t_color, edgecolor='k', alpha=alpha, label='Condition-ind.')
        bar_s = plt.bar(xtick, mEV_s, width, bottom=mEV_t, linewidth=line_width, color=s_color, edgecolor='k', alpha=alpha, label='Stimulus')
        bar_d = plt.bar(xtick, mEV_d, width, bottom=mEV_t+mEV_s, linewidth=line_width, color=d_color, edgecolor='k', alpha=alpha, label='Decision')
        bar_sd = plt.bar(xtick, mEV_sd, width, bottom=mEV_t+mEV_s+mEV_d, linewidth=line_width, color=sd_color, edgecolor='k', alpha=alpha, label='Interaction')

        #########################################################################################################################################################

        ax.legend(loc='upper right', frameon=True,fontsize=legend_fontsize,labelspacing=0,markerscale=legend_markerscale,borderaxespad=1)


    plt.tight_layout()
    if not os.path.exists('simulation/figures/N{}_P{}_S{}_D{}_T{}_s{}'.format(N, P, S, D, T, sparsity)):
        os.mkdir('simulation/figures/N{}_P{}_S{}_D{}_T{}_s{}'.format(N, P, S, D, T, sparsity))
    plt.savefig('simulation/figures/N{}_P{}_S{}_D{}_T{}_s{}/mEV.pdf'.format(N, P, S, D, T, sparsity, name2[method]),format='pdf',bbox_inches = 'tight')
    plt.show()

# draw_marginEVs(10, 100, 150, 6, 4, 0.8)
settings.traverse(draw_marginEVs)
```
